```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

preorders = pd.read_csv("../../../Project/Data/train_test_splits/backgroundData_backGroundTrainTestsplit.csv", sep=',')
```

Motivation: Maybe we want to know when in the 6 months period the user was buying the most? Maybe just at the beginning, or just at the end or over the whole period. (Idea: calculate expected date and standard deviation for this)

First: normalize every date to the day in the 6 months period (ranging from 0 to 6*30,25)

Expected Date: SUM OVER {BOUGHT PRODUCTS} WITH i [(i_BOUGHT_COUNT * i_NORMALIZED_DATE)] / COUNT(BOUGHT_PRODUCTS)
Variance: SUM OVER {BOUGHT PRODUCTS} WITH i [(i_BOUGHT_COUNT * (i_NORMALIZED_DATE - Expected Date)^2)] / (COUNT(BOUGHT_PRODUCTS)
Standard Deviation: Sqrt(Variance)

```python
preorders
```

```python
#Normalize the dates
# https://stackoverflow.com/questions/31036148/how-to-standardize-normalize-a-date-with-pandas-numpy
tmp = pd.DataFrame()
tmp["date"] = pd.to_datetime(preorders['date']).astype('int64')
min_date = tmp.date.min()
preorders['normalizedDate'] = (tmp.date - min_date) / 86400 / 1000000000 + 1.0
preorders.groupby(["normalizedDate"]).count()
```

```python
"""
normalize dates
get list of users
for user in users:
    make list of (normalized_date, summed up count of bought items)
    get avg (expected date) and std
    add the 2 features in the rows in orders with user
"""

avgDates = preorders.groupby(["userID"]).mean()["normalizedDate"]

orders = avgDates.to_frame()
orders["userID"] = orders.index
orders.index.name = ""
orders = orders.rename({"normalizedDate":"avgBuyDate"}, axis=1)
orders
```

```python
plt.hist(orders["avgBuyDate"])
plt.show()
```

```python

stdDates = preorders.groupby(["userID"]).std()["normalizedDate"]

orders2 = stdDates.to_frame()
orders2["userID"] = orders2.index
orders2.index.name = ""
orders2 = orders2.rename({"normalizedDate":"stdBuyDate"}, axis=1)
orders2
```

```python
orders = pd.merge(orders, orders2, on="userID")
#orders.drop("stdBuyDate_y", axis=1)
#orders.rename("stdBuyDate_x", axis=1)
orders
```

```python
plt.hist(orders["avgBuyDate"], alpha=0.5)
plt.hist(orders["stdBuyDate"], alpha=0.5)
plt.show()
```
